```python
def qdevtwo(x):
    if int(x) % 2 == 0:
        return 1
    else:
        return 0
        
pizzaz_by_province['dev2'] = pizzaz_by_province.quantity.apply(lambda x: qdevtwo(x))
```

```python
pizzaz_by_province['dev2'] = pizzaz_by_province.quantity.apply(qdevtwo)
```
